# Step 1: Tokenize the data, remove stop words and perform stemming or lemmatization.

```python
import nltk
import re
import pandas as pd
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

# 1. Download dataset NLTK yang diperlukan
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('punkt_tab')

# 2. Baca file CSV
df = pd.read_csv('semua_ulasan_hm.csv', engine='python', on_bad_lines='skip')

# Menampilkan 5 baris pertama untuk memastikan kolom 'content' ada
print("5 Baris pertama dataset:")
display(df.head())
print(f"Total baris yang dimuat: {len(df)}")

# 3. PERBAIKAN: Mengambil data dari kolom 'content'
if 'content' in df.columns:
    df = df.dropna(subset=['content'])

    # Baru setelah itu ambil datanya
    text_data = df['content'].astype(str).head(50).tolist()
else:
    print("Warning: Kolom 'content' tidak ditemukan! Cek kembali nama kolom di CSV kamu.")

# 4. Inisialisasi Lemmatizer dan Stopwords
lemmatizer = WordNetLemmatizer()
stop_words_set = set(stopwords.words('english')) # Ditaruh di luar loop biar cepat
corpus = []

# 5. Proses Pembersihan Teks
for text in text_data:
    # Tokenisasi kalimat
    sentences = nltk.sent_tokenize(text)
    for sent in sentences:
        # Hapus karakter selain huruf (A-Z)
        sent = re.sub('[^a-zA-Z]', ' ', sent)
        # Kecilkan semua huruf
        sent = sent.lower()
        # Pecah jadi kata-kata
        words = sent.split()

        # Lemmatization & Hapus Stopwords
        cleaned_words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words_set]

        # Gabungkan kembali dan masukkan ke corpus
        final_sent = ' '.join(cleaned_words)
        if final_sent: # Pastikan tidak memasukkan string kosong
            corpus.append(final_sent)

# 6. Tampilkan hasil
print(f"\nJumlah kalimat yang diproses dalam corpus: {len(corpus)}")
print("Isi Corpus (5 pertama):")
print(corpus[:5])
```

# Step 2: List all unique words

```python
print("Corpus:", corpus)
```

```python
# Finding unique words
all_words = []
for sentence in corpus:
    for word in sentence.split():
        all_words.append(word)

unique_words = list(set(all_words))
print("Unique words:", unique_words)
```

#Step 3: Create a dictionary with mapping of words to a number. This should now be sorted on frequency of occurrence in descending order.

Finding unique words and their frequencies

```python
from collections import Counter

all_words = []
for sentence in corpus:
    for word in sentence.split():
        all_words.append(word)

word_frequencies = Counter(all_words)

# Print the word counts
print("Word Frequencies:", word_frequencies)

# Sorting words by frequency (descending)
sorted_words = sorted(word_frequencies.items(), key=lambda item: item[1], reverse=True)

# Creating the dictionary mapping words to numbers
word_to_number = {}
for i, (word, frequency) in enumerate(sorted_words):
    word_to_number[word] = i + 1  # Assign numbers starting from 1

print("Word to Number Mapping (Sorted by Frequency):", word_to_number)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# Modify token_pattern to include single-character words
cv = CountVectorizer(token_pattern=r'(?u)\b\w+\b')
independentFeatures = cv.fit_transform(corpus).toarray()

print("Shape of independentFeatures:", independentFeatures.shape)
print("First 10 feature names (words):")
print(cv.get_feature_names_out()[:10])
independentFeatures
```
