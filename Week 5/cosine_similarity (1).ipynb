# Vector Space Model: Cosine Similarity

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
```

## 1. Import Dataset

```python
corpus = [
    "Tahun lalu, nilai transaksi Tokopedia dan TikTok Shop saat momen sahur melonjak naik lebih dari 10 kali lipat berkat kampanye Ramadan Ekstra Seru.",
    "Data industri menunjukkan jumlah transaksi e-commerce meningkat lebih dari 76%, dan jumlah konsumen naik tajam dibandingkan bulan biasa, mencerminkan peluang besar bagi pelaku usaha khususnya UMKM di platform e-commerce, seperti Tokopedia dan TikTok Shop.",
    "Bulan suci menghadirkan peluang positif bagi para pelaku industri untuk mengoptimalkan kinerja dan pertumbuhan. Tokopedia dan TikTok Shop mencatat, pesanan penjual yang yang berpartisipasi dalam kampanye Ramadan Ekstra Seru tahun lalu naik rata-rata lebih dari dua kali lipat."
]

vectorizer = CountVectorizer(vocabulary=['tiktok', 'naik'])
X = vectorizer.fit_transform(corpus)
```

## 2. Cosine Similarity

```python
# Membandingkan semua dokumen dengan Dokumen 0 (1 John)
cos_sim = cosine_similarity(X[0:1], X)

# Tampilkan Hasil
df = pd.DataFrame(X.toarray(), columns=['tiktok', 'naik'])
df['Cosine Sim to Doc 0'] = cos_sim.flatten()
print(df)
```

```python
import numpy as np
def length(v):
    return np.sqrt((v ** 2).sum())
def cos_dist(v, w):
    return 1 - (v * w).sum() / (length(v) * length(w))
cos_dist(tf.loc[0], tf.loc[1]), cos_dist(tf.loc[0], tf.loc[2])
```

```python
import pandas as pd
from collections import Counter

# Data dari slide
documents = [
    "Tiktok shop naik", # Doc 0
    "kenaikan tiktok shop",                # Doc 1
    "tiktok shop kenaikan" # Doc 2
]

# Membuat DataFrame TF secara manual
tf = pd.DataFrame(
    [Counter(doc.split()) for doc in documents],
).fillna(0).astype(int)

# Menampilkan matriks
print("Matriks Term Frequency (TF):")
print(tf)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

full_cos_sim_matrix = cosine_similarity(tf)

# Display the full matrix as a DataFrame for better readability
cos_sim_df = pd.DataFrame(
    full_cos_sim_matrix,
    index=[f'Doc {i}' for i in range(len(documents))],
    columns=[f'Doc {i}' for i in range(len(documents))]
 )
print("Full Cosine Similarity Matrix:")
print(cos_sim_df)
```
